```python
with open("sentences.txt") as file_obj:
    data_list = list(map(lambda x: x.lower(), map(lambda x: x.strip(), file_obj.readlines())))
data_list
len(data_list)
```

```python
import re
data_list = list(map(lambda t: re.split('[^a-z]', t), data_list))  # регулярное выражение, считающее разделителем любой символ, не являющийся буквой
data_list = [list(filter(lambda word: word != "", sentence)) for sentence in data_list]
data_list
```

```python
words = dict()
for sentence in data_list:
    for word in sentence:
        words[word] = words.get(word, 0) + 1
words
len(words)
```
